BTC_USD profits and losses

## Importing libraries 

```python
import yfinance as yf
import numpy as np  
import pandas as pd
from matplotlib import pyplot as plt
```

## Downloading and reading dataframe

```python
dataframe = yf.download("BTC-USD", start="2020-01-01", end="2021-01-01", interval="1d")
dataframe
```

## BTC-USD benefit

```python
dataframe["Benefit"] = dataframe["Close"] - dataframe["Open"]
dataframe["Profit"] = np.where(dataframe["Benefit"] > 0, "Profit", "Loss")
```

## Showing BTC-USD prices from 2020-2021 in a plot

```python
dataframe[['Close', 'Benefit']].plot(figsize=(12,6))
plt.title("BTC-USD Close Price and Daily Benefit")
plt.xlabel("Date")
plt.ylabel("Price (USD) / Benefit (USD)")
plt.legend(['Close Price', 'Daily Benefit'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
```
